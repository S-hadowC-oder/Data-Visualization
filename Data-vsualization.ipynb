```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate   x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid
x, y = np.meshgrid(x, y)

# Calculate z values
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame for visualization (optional)
data = {
    'x': x.flatten(),
    'y': y.flatten(),
    'z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D figure and axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', alpha=0.8)

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot of sin(sqrt(x^2 + y^2))')

# Add a colorbar
fig.colorbar(surf, shrink=0.5, aspect=5)

# Show the plot
plt.show()
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show
import math  # For sine function

# Generate x-values from 0 to 10
x = list(range(11))

# Calculate y-values as sin(x)
y = [math.sin(i) for i in x]

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)

show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import random

# Generate random data (replace with your data source)
x = list(range(11))  # Generate 11 x-values
y = [random.random() for _ in x]  # Generate random y-values
sizes = [random.randint(10, 20) for _ in x]  # Random sizes for markers
colors = [f"rgb({random.randint(50, 200)}, {random.randint(50, 200)}, {random.randint(50, 200)})" for _ in x]  # Random colors

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new plot with a title and axis labels
p = figure(title="Scatter Plot", x_axis_label="X", y_axis_label="Y")

# Add circle renderer with color mapper and size mapping
p.circle(x="x", y="y", source=source, size="sizes", color="colors", fill_alpha=0.6)

show(p)
```

```python
 # 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]
from bokeh.plotting import figure, show

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruit", y_axis_label="Count")

# Add a vbar renderer with a specific color
p.vbar(x="fruits", top="counts", width=0.5, color="skyblue", source=dict(fruits=fruits, counts=counts))

# Display the plot
show(p)
```

```python
# # 4. Create a Bokeh histogram to visualize the distribution of the given data.

from bokeh.plotting import figure, show

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
# Use the provided data and histogram results
hist, edges = data_hist, edges  # Assuming data_hist and edges are defined elsewhere

# Create a new plot with a title and axis labels
p = figure(title="Histogram of Random Data", x_axis_label="Value", y_axis_label="Frequency")

# Add quad renderer to visualize bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="white")

# Display the plot
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource

# Prepare the output to display in a Jupyter notebook
output_notebook()

# Generate the dataset
data_heatmap = np.random.rand(10, 10)  # 10x10 random values
x = np.linspace(0, 1, 10)  # x-coordinates
y = np.linspace(0, 1, 10)  # y-coordinates
xx, yy = np.meshgrid(x, y)  # Create a meshgrid

# Flatten the arrays for Bokeh
data_flat = data_heatmap.flatten()
x_flat = xx.flatten()
y_flat = yy.flatten()

# Create a ColumnDataSource for the heatmap
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

# Create a figure
p = figure(title="Heatmap", 
           x_axis_label='X-axis', 
           y_axis_label='Y-axis', 
           tools="", 
           width=400,  # Use width instead of plot_width
           height=400)  # Use height instead of plot_height

# Define a color mapper
color_mapper = linear_cmap(field_name='value', palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Create a heatmap using rectangles
p.rect(x='x', y='y', width=0.1, height=0.1, 
        source=source,
        line_color=None, fill_color=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show theplot
show(p)
```
